```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import spacy
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bbc_news = pd.read_csv('E:/Vaibhav/Study_CS/365DataScience/bbc_news.csv')
bbc_news.head()
```

```python
bbc_news.info()
```

```python
titles = pd.DataFrame(bbc_news['title'])
titles
```

## Clean data 

```python
titles['lowercase'] = titles['title'].str.lower()
titles
```

```python
##stop word removal
en_stopwords = stopwords.words('english')
titles['no_stopwords'] = titles['lowercase'].apply(lambda x: ' '.join([word for word in x.split() if word not in (en_stopwords)]))
```

```python
##punctation removal
titles['no_stopwords_no_punct'] = titles.apply(lambda x: re.sub(r"[^\w\s]"," ",x['no_stopwords']),axis=1)
```

```python
##tokenize
titles['tokens_raw'] = titles.apply(lambda x: word_tokenize(x['title']),axis=1)
titles['tokens_clean'] = titles.apply(lambda x: word_tokenize(x['no_stopwords_no_punct']),axis=1)
```

```python
##lemmatizing
lemmatizer =  WordNetLemmatizer()
titles['token_clean_lemmatized'] = titles['tokens_clean'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
titles.head()
```

```python
token_raw_list = sum(titles['tokens_raw'], [])
token_clean_list = sum(titles['token_clean_lemmatized'], [] )
```

```python
token_raw_list
```

```python
token_clean_list
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
spacy_doc = nlp(' '.join(token_raw_list))
```

```python
pos_df = pd.DataFrame(columns=['token','pos_tag'])
```

```python
for token in spacy_doc:
    pos_df = pd.concat([pos_df, pd.DataFrame({'token':[token.text], 'pos_tag':[token.pos_]})],ignore_index=True)
```

```python
pos_df.head(10)
```

```python
pos_df_counts = pos_df.groupby(['token','pos_tag']).size().reset_index(name='counts').sort_values(by="counts",ascending=False)
```

```python
pos_df_counts
```

```python
nouns = pos_df_counts[pos_df_counts.pos_tag == 'NOUN'][:10]
nouns
```

## NAMED ENTITY RECOGNITION

```python
ner_df = pd.DataFrame(columns=['token', 'ner_tag'])

for token in spacy_doc.ents:
    if pd.isna(token.label_) is False:
        new_row = pd.DataFrame({'token': [token.text], 'ner_tag': [token.label_]})
        ner_df = pd.concat([ner_df, new_row], ignore_index=True)
```

```python
ner_df.head()
```

```python
ner_df_counts =  ner_df.groupby(['token','ner_tag']).size().reset_index(name='counts').sort_values(by="counts",ascending=False)
ner_df_counts
```

```python
people = ner_df_counts[ner_df_counts.ner_tag == 'PERSON'][:10]
people
```
